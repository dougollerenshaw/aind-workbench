```python
# Let's verify the folder structure and count both total and unique subject IDs
import os
import re

# Get all folder names
folders = [f for f in os.listdir('barseq_metadata') if not f.startswith('.')]
print(f"Number of folders: {len(folders)}")
print("Folders:")
for folder in folders:
    print(f"  {folder}")

# Extract all subject IDs from folder names (after "TO_")
all_subject_ids = []
for folder in folders:
    # Extract the part after "TO_" and split by underscores
    if "_TO_" in folder:
        subjects_part = folder.split("_TO_")[1]
        subjects = subjects_part.split("_")
        all_subject_ids.extend(subjects)
        print(f"\n{folder}:")
        print(f"  Subjects: {subjects}")

print(f"\nTotal subject ID mentions: {len(all_subject_ids)}")
print(f"Unique subject IDs: {len(set(all_subject_ids))}")
print(f"Unique subjects: {sorted(set(all_subject_ids))}")
```

```python
# Let's see which subject IDs appear in multiple folders
from collections import Counter

# Count how many times each subject ID appears
subject_counts = Counter(all_subject_ids)

print("Subject ID frequency across folders:")
for subject_id, count in sorted(subject_counts.items()):
    if count > 1:
        print(f"  {subject_id}: appears {count} times ⭐")
    else:
        print(f"  {subject_id}: appears {count} time")

# Show which specific folders each repeated subject appears in
print("\nRepeated subjects and their folders:")
repeated_subjects = [sid for sid, count in subject_counts.items() if count > 1]

for subject_id in sorted(repeated_subjects):
    folders_with_subject = []
    for folder in folders:
        if "_TO_" in folder:
            subjects_in_folder = folder.split("_TO_")[1].split("_")
            if subject_id in subjects_in_folder:
                folders_with_subject.append(folder)
    
    print(f"\n{subject_id} appears in:")
    for folder in folders_with_subject:
        print(f"  - {folder}")

print(f"\nSummary:")
print(f"Total subject mentions: {len(all_subject_ids)} (4 subjects × {len(folders)} folders)")
print(f"Unique subjects: {len(set(all_subject_ids))}")
print(f"Repeated subjects: {len(repeated_subjects)}")
```
