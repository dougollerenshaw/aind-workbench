```python
from aind_data_access_api.document_db import MetadataDbClient
import pandas as pd

# Configure pandas to display all columns
pd.set_option('display.max_columns', None)
```

```python
# Initialize the client
client = MetadataDbClient(
    host="api.allenneuraldynamics.org",
    database="metadata_index",
    collection="data_assets",
)
```

```python
mouse_id = [
    "725231",
    "731907", 
    "744119",
    "744117",
    "746046",
    "746045",
    "746043",
    "746041",
    "751023",
    "751024",
    "751035", 
    "751019",
    "751017",
    "755071",
    "755073",
    "755072",
    "755069",
    "755790",
    "762196",
    "762199"
]
```

```python
query = {"subject.subject_id": {"$in": mouse_id}}

results = client.retrieve_docdb_records(
    filter_query=query
)

len(results)
```

```python
df = pd.json_normalize(results, sep='.', max_level=None)
df.head()
```

```python
# Group by subject_id and get unique data levels with their counts
data_level_counts = df.groupby('subject.subject_id')['data_description.data_level'].value_counts().unstack(fill_value=0)

# Add a total column to show total records per subject
data_level_counts['total'] = data_level_counts.sum(axis=1)

# Sort by subject ID for easier reading
data_level_counts.sort_index(inplace=True)

# Display the results
print("Number of records per data level for each subject:")
display(data_level_counts)
```
